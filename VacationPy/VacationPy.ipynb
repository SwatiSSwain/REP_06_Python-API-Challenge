```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
#load csv into dataframe
load_csv = "../output_data/cities.csv"
cities_df = pd.read_csv(load_csv)
cities_df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Store latitude and longitude in locations
locations1 = cities_df[["Lat", "Lng"]]

# Fill NaN values and convert to float
humidity = cities_df["Humidity"].astype(float)

#For max intensity in the heat map, try setting it to the highest humidity found in the data set.
max_humidity = max(humidity)
new_humidity = [max_humidity for x in humidity]

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations1, weights=new_humidity, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
#Create new df with below criteria:
#* A max temperature lower than 80 degrees but higher than 70.
#* Wind speed less than 10 mph.
#* Zero cloudiness.

ideal_weather_df=cities_df.loc[(cities_df['Max Temp']>70) & (cities_df['Max Temp']<80) 
                               & (cities_df['Wind Speed'] < 10) 
                               & (cities_df['Cloudiness']==0)]

ideal_weather_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#Store ideal_weather_df into variable named hotel_df.
hotel_df=ideal_weather_df.copy()

#Add a "Hotel Name" column to the DataFrame.
hotel_df['Hotel Name'] = ""

#Display Hotel df
hotel_df
```

```python
#Parametize for API call
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "radius":5000, #Set parameters to search for hotels with 5000 meters
    "type": "lodging",
    "key": g_key,
}
# use iterrows to iterate through pandas dataframe to get hotels within 5000 meters of the ideal cities
for index, row in hotel_df.iterrows():

    # get latitude & longitude from df
    lat = row['Lat']
    lng = row['Lng']

    # add location to params dict
    params['location'] = f"{lat},{lng}"

    # assemble url and make API request
    print(f"Retrieving Results for {row['City']}.")
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    
    try:
        print(f"Closest hotel for {row['City']} is {results[0]['name']}.")
        #Store first Hotel Name in df
        hotel_df.loc[index, 'Hotel Name'] = results[0]['name']
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")

#Data clea up - Set NAN as Hotel Name where no hotel is present within 5000 m radius and drop those records
hotel_df['Hotel Name'] = hotel_df['Hotel Name'].replace('', np.nan)
hotel_df = hotel_df.dropna()

#Display final df
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map

#Hotel Marker
hotel_markers = gmaps.marker_layer(locations, info_box_content=hotel_info)

#Add Marker Layer
fig.add_layer(hotel_markers)

# Display figure
fig
```
